```python
import pandas as pd
import quantstats as qs
import yfinance as yf
import pytz
import datetime

qs.extend_pandas()
```

## Organización de data

```python
sample = 'Registro_de_operaciones_Roboforex_Febrero2023.xlsx'
```

```python
df = pd.read_excel(sample, index_col=0)
df
```

```python
df['pct_bal'] = df['Balance'].pct_change()
```

```python
df.reset_index(inplace=True)
df.dtypes
```

```python
df
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df.set_index('datetime', inplace=True)
```

```python
returns = df['pct_bal']
returns
```

```python
resampled = returns.resample('D').sum()
resampled.index = resampled.index.tz_localize(None)
resampled
```

## Aplico con columna Balance

```python
qs.reports.html(resampled, 'EURUSD=X', output='DA_vs_EURUSD.html', title='DA_vs_EURUSD')
```
